```python
#2.30分貝的聲壓會是50分貝的幾倍?
#公式移項過後可以得到 V1 = ?
#請寫下程式
import numpy as np 
Gdb=30
V0=20
V_30= 10**(Gdb/20)*V0
Gdb=50
V_50= 10**(Gdb/20)*V0
print(V_30/V_50)
```
