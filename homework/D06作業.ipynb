作業目標<br>
讀取存取陣列資料<br>
作業重點<br>
多陣列存一起需要存成npz，讀取須注意npz中有多個陣列

題目:<br>
1. 將下兩列array存成npz檔<br>
array1 = np.array(range(30))<br>
array2 = np.array([2,3,5])<br>
2. 讀取剛剛的npz檔，加入下列array一起存成新的npz檔


```python
#. 將下兩列array存成npz檔
import numpy as np
array1 = np.array(range(30))
array2 = np.array([2,3,5])
np.savez('saved_array.npz',array1=array1,array2=array2)
#2. 讀取剛剛的npz檔，加入下列array一起存成新的npz檔
Saved_file=np.load('saved_array.npz')
array3=[1,2,3]
np.savez('saved_array.npz',array1=array1,array2=array2,array3=array3)


Saved_file=np.load('saved_array.npz')
print(Saved_file.files)
print(Saved_file['array3'])
```

```python
#. 將下兩列array存成npz檔
import numpy as np
array1 = np.array(range(30))
array2 = np.array([2,3,5])
with open('saved_array.npz', 'wb') as sa:
    np.savez(sa,array1=array1,array2=array2)
#2. 讀取剛剛的npz檔，加入下列array一起存成新的npz檔
Saved_file=np.load('saved_array.npz')
array3=[1,2,3]
with open('saved_array.npz', 'wb') as sa:
    np.savez(sa,array1=array1,array2=array2,array3=array3)
Saved_file=np.load('saved_array.npz')
Saved_file.files
```
